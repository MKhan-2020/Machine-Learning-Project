# Part - II Data Manipulation

Subtask #2: Estimate the offset of the values of "MOS" to UTC. Store them in a column called "OffsetUTC".

```python
sales_data['MOS'].iloc[0]
```

```python
sales_data['MOS'].iloc[0].tz_localize(sales_data['TimeZone'].iloc[0])
```

```python
lst_mos_timestamps = []
lst_mos_timestamps
```

```python
for i in range(len(sales_data)):
    lst_mos_timestamps.append(pd.to_datetime(sales_data['MOS'].iloc[i].tz_localize(sales_data['TimeZone'].iloc[i], ambiguous = 'raise', nonexistent = 'raise')))
```

```python
lst_mos_timestamps[:10]
```

```python
lst_mos_timestamps
```

```python
lst_mos_timestamps[0].strftime('%Y-%m-%d %H:%M:%S')
```

```python
lst_mos_timestamps[0].strftime('%z')
```

```python
float(lst_mos_timestamps[0].strftime('%z'))/100
```

```python
lst_mos_timestamps[0].date()
```

```python
lst_mos_timestamps[0].time()
```

```python
lst_mos_timestamps[0].utcoffset()
```

```python
lst_mos_timestamps[0].utcoffset().total_seconds() / 3600
```

```python
sales_data['OffsetUTC'] = [lst_mos_timestamps[i].utcoffset().total_seconds() / 3600 for i in range(len(sales_data))]

sales_data.head()
```

```python
column_reorder = ['RetailerCountry', 'MOS', 'TimeZone', 'OffsetUTC', 'RetailerType', 'Product', 'Sales Revenue ($)']

sales_data = sales_data[column_reorder]
sales_data.head()
```

```python
sales_data.groupby(['TimeZone', 'OffsetUTC']).count()
```

```python
sales_data[sales_data['TimeZone'] == 'CET'].sort_values('OffsetUTC')
```

# Part - III Data Analysis

Subtask #3: Order all sales according to a reconciled UTC-equivalent of the moment of sales ('MOS') and analyze the data.

```python
sales_data.sort_values('MOS')
```

```python
sales_data['MOS-UTC'] = [pd.to_datetime(sales_data['MOS'].iloc[i]).tz_localize(sales_data['TimeZone'].iloc[i]).tz_convert('UTC') 
                        for i in range(len(sales_data))]
                           
                                                                             
```

```python
sales_data['MOS-UTC'].head()
```

```python
column_reorder = ['RetailerCountry', 'MOS-UTC', 'MOS', 'TimeZone', 'OffsetUTC', 'RetailerType', 'Product', 'Sales Revenue ($)']

sales_data = sales_data[column_reorder]
sales_data.head()
```

```python
sales_data.sort_values(['MOS-UTC'])
```

```python
sales_data.groupby('TimeZone').mean()
```

```python
sales_data.groupby('TimeZone').count()
```

```python
sales_data.groupby(['TimeZone','OffsetUTC']).count()
```

# Part - IV Data Visualization

Subtask #4: Obtain Statistics and Visualizations that will respond to the Main Task and further improve the analysis of data

```python
sales_data.head()
```

```python
sales_data_viz = sales_data.reset_index()
sales_data_viz.head()
```

```python
sales_data_viz = sales_data_viz.set_index('MOS-UTC').sort_index()

sales_data_viz.head()
```

```python
sales_data_viz['OffsetUTC'].plot(figsize=(10,6))
```

```python
plt.scatter(sales_data_viz['OffsetUTC'], sales_data_viz['RetailerType'])
```

```python
sales_data_viz.groupby('TimeZone').sum()
```

```python
total_sales_revenue_by_time_Zone = sales_data_viz.groupby('TimeZone').sum()
total_sales_revenue_by_time_Zone
```

```python
plt.bar(total_sales_revenue_by_time_Zone.index, 
        height = total_sales_revenue_by_time_Zone["Sales Revenue ($)"])
```

```python
plt.figure(figsize=(7,6))

plt.bar(total_sales_revenue_by_time_Zone.index, 
        height = total_sales_revenue_by_time_Zone["Sales Revenue ($)"], 
        color = '#108A99')

plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Total Sales Revenue by Time Zone", fontsize = 18, fontweight = 'bold')
plt.ylabel("Sales Revenue ($)", fontsize = 13)


plt.show()
```

```python
sales_data_viz.groupby(['TimeZone', 'OffsetUTC']).mean()
```

```python
sales_data_viz.groupby(['TimeZone']).mean()
```

```python
sales_data_viz.groupby(['TimeZone']).mean().sort_values("Sales Revenue ($)")
```

```python
average_sales_data_by_TimeZone = sales_data_viz.groupby(['TimeZone']).mean().sort_values("Sales Revenue ($)", ascending = False)

average_sales_data_by_TimeZone
```

```python
plt.bar(average_sales_data_by_TimeZone.index, 
        height = average_sales_data_by_TimeZone["Sales Revenue ($)"])
```

```python
plt.figure(figsize=(7,6))

plt.bar(average_sales_data_by_TimeZone.index, 
        height = average_sales_data_by_TimeZone["Sales Revenue ($)"], 
        color = ['green', 'blue', 'red', 'darkblue', 'yellow', 'purple'])

plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Average Sales Revenue by Time Zone", fontsize = 18, fontweight = 'bold')
plt.ylabel("Average Sales Revenue ($)", fontsize = 13)


plt.show()
```
